```python
def remove_quotes_from_list(lst):
    return [element.replace("'", "") for element in lst]
```

```python
book['Genres'] = remove_quotes_from_list(book['Genres'])
```

```python
book
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
book['Genres'] = collapse(book['Genres'])
```

```python
book
```

```python
book['Author']
```

```python
book = book[['Book','Description','Genres','Author']]
```

```python
book
```

```python
book['tags'] = book['Description'] + book['Genres'] + book['Author']
```

```python
book
```

```python
new = book.drop(columns=['Description','Genres','Author'])
```

```python
new
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
# Assuming 'new' is your DataFrame and 'tags' is the column containing tags
new['tags'] = new['tags'].fillna('')  # Replace NaN values with an empty string

# Now, you can use CountVectorizer to transform the 'tags' column
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(book_name):
    if book_name not in new['Book'].values:
        print(f"Book '{book_name}' not found in the dataset.")
        return

    index = new[new['Book'] == book_name].index[0]  # Use index[0] instead of index[1]
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].Book)
```

```python
recommend('The Diary of a Young Girl')
```
