```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df = pd.read_csv('moviedb.csv')
sex = df.drop(['FILM','RT_norm','RT_user_norm','Metacritic_norm','Metacritic_user_nom','IMDB_norm','RT_norm_round','RT_user_norm_round','Metacritic_norm_round','Metacritic_user_norm_round','IMDB_norm_round','Metacritic_user_vote_count','IMDB_user_vote_count','Fandango_votes','Fandango_Difference'], axis=1)

# Calculate the correlation matrix
correlation_matrix = sex.corr()
sthan_corr = sex.corr(method='kendall')

# Set up the matplotlib figure
plt.figure(figsize=(14, 8))

# Draw the heatmap using seaborn
sb.heatmap(sthan_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

# Set the title of the plot
plt.title('Correlation Matrix')

# Show the plot
plt.show()
```
